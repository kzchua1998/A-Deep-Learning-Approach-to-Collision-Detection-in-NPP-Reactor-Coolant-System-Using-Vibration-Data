### Vanilla 1D Convolutional Neural Network

```python
scaler = 1
n_feature_map = scaler*[16, 32, 64]
defaultConv1D = partial(tf.keras.layers.Conv1D,kernel_size=3,
                                               kernel_initializer='he_normal')
# With Regularizers
'''alpha = 0.01
regularizer = regularizers.l1(0.01)
regularizer = regularizers.l2(0.01)

defaultConv1D = partial(tf.keras.layers.Conv1D,kernel_size=3,
                                               kernel_initializer='he_normal',
                                               kernel_regularizer=regularizer, 
                                               bias_regularizer=regularizer)

defaultConv1D = partial(tf.keras.layers.Conv1D,kernel_size=3,
                                               kernel_initializer='he_normal', 
                                               activity_regularizer=regularizer)'''
input_shape=(375, 4)
model = tf.keras.Sequential()

# Conv1
model.add(defaultConv1D(input_shape=input_shape,filters=scaler*n_feature_map[0],name='Conv1a'))
model.add(tf.keras.layers.BatchNormalization(momentum=0.9))
model.add(tf.keras.layers.Activation('relu'))
model.add(defaultConv1D(filters=scaler*n_feature_map[0],name='Conv1b'))
model.add(tf.keras.layers.BatchNormalization(momentum=0.9))
model.add(tf.keras.layers.Activation('relu'))
model.add(tf.keras.layers.MaxPooling1D(pool_size=2))

# Conv2
model.add(defaultConv1D(filters=scaler*n_feature_map[1],name='Conv2a'))
model.add(tf.keras.layers.BatchNormalization(momentum=0.9))
model.add(tf.keras.layers.Activation('relu'))
model.add(defaultConv1D(filters=scaler*n_feature_map[1],name='Conv2b'))
model.add(tf.keras.layers.BatchNormalization(momentum=0.9))
model.add(tf.keras.layers.Activation('relu'))
model.add(tf.keras.layers.MaxPooling1D(pool_size=2))

# Conv3
model.add(defaultConv1D(filters=scaler*n_feature_map[2],name='Conv3a'))
model.add(tf.keras.layers.BatchNormalization(momentum=0.9))
model.add(tf.keras.layers.Activation('relu'))
model.add(defaultConv1D(filters=scaler*n_feature_map[2],name='Conv3b'))
model.add(tf.keras.layers.BatchNormalization(momentum=0.9))
model.add(tf.keras.layers.Activation('relu'))
model.add(tf.keras.layers.MaxPooling1D(pool_size=2))

# Top Layer
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(units=50,kernel_initializer='he_normal'))
model.add(tf.keras.layers.BatchNormalization(momentum=0.9))
model.add(tf.keras.layers.Activation('relu'))
model.add(tf.keras.layers.Dropout(0.5))
model.add(tf.keras.layers.Dense(units=1))
    
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4),
              loss='mean_squared_error',
              metrics= tf.keras.metrics.MeanSquaredError())
```

### Wide & Deep 1D Convolutional Neural Network

```python
scaler = 1
n_feature_map = scaler*[16, 32, 64]
defaultConv1D = partial(tf.keras.layers.Conv1D,kernel_size=3,
                                               kernel_initializer='he_normal')
# With Regularizers
'''alpha = 0.01
regularizer = regularizers.l1(0.01)
regularizer = regularizers.l2(0.01)

defaultConv1D = partial(tf.keras.layers.Conv1D,kernel_size=3,
                                               kernel_initializer='he_normal',
                                               kernel_regularizer=regularizer, 
                                               bias_regularizer=regularizer)

defaultConv1D = partial(tf.keras.layers.Conv1D,kernel_size=3,
                                               kernel_initializer='he_normal', 
                                               activity_regularizer=regularizer)'''
input_shape=(375, 4)
model = tf.keras.Sequential()
input_layer = tf.keras.layers.Input(shape=input_shape)
# Conv1
x = defaultConv1D(filters=scaler*n_feature_map[0],name='Conv1a')(input_layer)
x = tf.keras.layers.BatchNormalization(momentum=0.9)(x)
x = tf.keras.layers.Activation('relu')(x)
x = defaultConv1D(filters=scaler*n_feature_map[0],name='Conv1b')(x)
x = tf.keras.layers.BatchNormalization(momentum=0.9)(x)
x = tf.keras.layers.Activation('relu')(x)
x = tf.keras.layers.MaxPooling1D(pool_size=2)(x)

# Conv2
x = defaultConv1D(filters=scaler*n_feature_map[1],name='Conv2a')(x)
x = tf.keras.layers.BatchNormalization(momentum=0.9)(x)
x = tf.keras.layers.Activation('relu')(x)
x = defaultConv1D(filters=scaler*n_feature_map[1],name='Conv2b')(x)
x = tf.keras.layers.BatchNormalization(momentum=0.9)(x)
x = tf.keras.layers.Activation('relu')(x)
x = tf.keras.layers.MaxPooling1D(pool_size=2)(x)

# Conv3
x = defaultConv1D(filters=scaler*n_feature_map[2],name='Conv3a')(x)
x = tf.keras.layers.BatchNormalization(momentum=0.9)(x)
x = tf.keras.layers.Activation('relu')(x)
x = defaultConv1D(filters=scaler*n_feature_map[2],name='Conv3b')(x)
x = tf.keras.layers.BatchNormalization(momentum=0.9)(x)
x = tf.keras.layers.Activation('relu')(x)
x = tf.keras.layers.MaxPooling1D(pool_size=2)(x)

# Input Flatten Layer

input_flatten = tf.keras.layers.Flatten()(input_layer)

# Top Layer
flatten = tf.keras.layers.Flatten()(x)
concat = tf.keras.layers.Concatenate()([input_flatten,flatten])
x = tf.keras.layers.Dense(units=1024,kernel_initializer='he_normal)(concat)
x = tf.keras.layers.BatchNormalization(momentum=0.9)(x)
x = tf.keras.layers.Activation('relu')(x)

x = tf.keras.layers.Dense(units=256,kernel_initializer='he_normal')(x)
x = tf.keras.layers.BatchNormalization(momentum=0.9)(x)
x = tf.keras.layers.Activation('relu')(x)

x = tf.keras.layers.Dense(units=64,kernel_initializer='he_normal')(x)
x = tf.keras.layers.BatchNormalization(momentum=0.9)(x)
x = tf.keras.layers.Activation('relu')(x)

x = tf.keras.layers.Dropout(0.5)(x)
output = tf.keras.layers.Dense(units=4)(x)

model = tf.keras.models.Model(inputs=[input_layer],outputs=[output])
    
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4),
              loss='mean_squared_error',
              metrics= tf.keras.metrics.MeanSquaredError())
```

# Hyperparameter Tuning using Keras Tuner

```python
def PatchNet_HP(hp):
    scaler = hp.Int('scaler',min_value=1,max_value=10,step=1)
    defaultConv1D = partial(tf.keras.layers.Conv1D,kernel_size=3,
                                                   kernel_initializer='he_normal')
    momentums=hp.Choice('momentum',[0.9,0.99,0.999])
    learning_rate = hp.Choice('learning_rate',[0.1,0.01,0.001])
    
    model = tf.keras.Sequential()
    model.add(defaultConv1D(filters=3*scaler))
    model.add(tf.keras.layers.BatchNormalization(momentum=momentums))
    model.add(tf.keras.layers.Activation('selu'))
    model.add(tf.keras.layers.MaxPooling1D(pool_size=2))
    model.add(defaultConv1D(filters=4*scaler))
    model.add(tf.keras.layers.BatchNormalization(momentum=momentums))
    model.add(tf.keras.layers.Activation('selu'))
    model.add(tf.keras.layers.MaxPooling1D(pool_size=2))
    model.add(defaultConv1D(filters=5*scaler))
    model.add(tf.keras.layers.BatchNormalization(momentum=momentums))
    model.add(tf.keras.layers.Activation('selu'))
    model.add(tf.keras.layers.MaxPooling2D(pool_size=2))
    model.add(defaultConv1D(filters=6*scaler))
    model.add(tf.keras.layers.BatchNormalization(momentum=momentums))
    model.add(tf.keras.layers.Activation('selu'))
    model.add(tf.keras.layers.MaxPooling1D(pool_size=2))
    model.add(tf.keras.layers.Flatten())
    model.add(tf.keras.layers.Dense(units=50,kernel_initializer=tf.keras.initializers.lecun_normal))
    model.add(tf.keras.layers.BatchNormalization(momentum=momentums))
    model.add(tf.keras.layers.Activation('selu'))
    model.add(tf.keras.layers.Dropout(0.5))
    model.add(tf.keras.layers.Dense(units=4,activation='softmax'))

    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=learning_rate),
                  loss='mean_squared_error',
                  metrics=['mean_squared_error'])
    return model
```

```python
tuner1 = BayesianOptimization(PatchNet_HP,
                              objective='val_mean_squared_error',
                              max_trials=20,
                              directory='D:/Users/Admin/Desktop/raw/weights')
```
